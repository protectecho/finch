```python
% matplotlib inline
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('img/input.jpg')

h, w = image.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D((w/2, h/2), 90, 1.0) # divide by two to rotate the image around its centre

rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

plt.imshow(rotated_image)
```

Another method for simple rotations that uses the cv2.transpose function

```python
% matplotlib inline
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('img/input.jpg')

rotated_image = cv2.transpose(img)

plt.imshow(rotated_image)
```
