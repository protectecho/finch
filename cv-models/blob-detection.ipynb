#### cv2.drawKeypoints(input image, keypoints, blank_output_array, color, flags)

```python
% matplotlib inline
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('img/sunflowers.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
image = cv2.imread('img/sunflowers.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
detector = cv2.SimpleBlobDetector_create() # set up detector
keypoints = detector.detect(image) # detect blobs
blobs = cv2.drawKeypoints(image, keypoints, np.zeros((1,1)), (0,255,255), cv2.DRAW_MATCHES_FLAGS_DEFAULT)
plt.imshow(blobs)
```
